```python
import pandas as pd

df = pd.DataFrame({'key':['A','B','C','A','B','C','A','B','C'],
                  'data':[0,5,10,5,10,15,10,15,20]})
df
```

```python
for key in ['A','B','C']:
    print (key,df[df['key'] == key].sum())
```

```python
df.groupby('key').sum()
```

```python
import numpy as np
df.groupby('key').aggregate(np.mean)
df.groupby('key').aggregate(np.sum)
```
